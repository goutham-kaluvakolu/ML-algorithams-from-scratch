```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

## reading data from the files
columns = ['Height', 'diameter','Weight', 'Hue','class']
data_df = pd.read_csv("Program_Data.csv", names = columns, header=0)

y_data = pd.get_dummies(data_df['class']).values.astype(float)
x_data = data_df[['Height', 'diameter','Weight', 'Hue']].values.astype(float)

ep = 20

def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

def get_gradient(y,x,w):
    y_pred = softmax(np.dot(x, w))
    return np.dot(x.T, y_pred - y)

def model_train(x_in, y_in, lr, iters):
    x = np.array(x_in)
    y = np.array(y_in)

    # convert string labels to one-hot encoded vectors
    if y.ndim == 1:
        y = pd.get_dummies(y_in).values.astype(float)
        
    w = np.random.normal(0, 0.01, size=(x.shape[1], y.shape[1]))
    prev_error = float('inf')
    for i in range(iters):
        w_d = get_gradient(y, x, w)
        y_pred = softmax(np.dot(x, w))
        cur_error = -np.mean(y * np.log(y_pred))

        if cur_error > prev_error:
            break

        w = w - lr * w_d
        prev_error = cur_error

    return w

# Leave-One-Out cross-validation
n = x_data.shape[0]
accs = []
for i in range(n):
    x_train = np.delete(x_data, i, axis=0)
    y_train = np.delete(y_data, i, axis=0)
    x_val = x_data[i:i+1,:]
    y_val = y_data[i:i+1,:]

    x_train_vec = np.hstack((x_train, np.ones((x_train.shape[0], 1))))
    weights = model_train(x_train_vec, y_train, 0.00005, 1000)

    x_val_vec = np.hstack((x_val, np.ones((x_val.shape[0], 1))))
    y_pred = softmax(np.dot(x_val_vec, weights))
    
    y_pred_label = np.argmax(y_pred, axis=1)
    y_val_label = np.argmax(y_val, axis=1)
    
    accs.append(np.sum(y_pred_label == y_val_label) / 1.0)

print("Accuracy: ", np.mean(accs))

# Train on the entire training set and make predictions on the test set
x_train_vec = np.hstack((x_train, np.ones((x_train.shape[0], 1))))
weights = model_train(x_train_vec, y_train, 0.00005, 1000)

x_test_vec = np.hstack((x_test, np.ones((x_test.shape[0], 1))))
# y_pred_test = softmax(np.dot(x_test_vec, weights))
# print(y_pred_test)

x_data = data_df[['Height', 'diameter', 'Weight']].values.astype(float)
y_data = pd.get_dummies(data_df['class']).values.astype(float)

n = x_data.shape[0]
accs = []
for i in range(n):
    x_train = np.delete(x_data, i, axis=0)
    y_train = np.delete(y_data, i, axis=0)
    x_val = x_data[i:i+1,:]
    y_val = y_data[i:i+1,:]

    x_train_vec = np.hstack((x_train, np.ones((x_train.shape[0], 1))))
    weights = model_train(x_train_vec, y_train, 0.00005, 1000)

    x_val_vec = np.hstack((x_val, np.ones((x_val.shape[0], 1))))
    y_pred = softmax(np.dot(x_val_vec, weights))
    
    y_pred_label = np.argmax(y_pred, axis=1)
    y_val_label = np.argmax(y_val, axis=1)
    
    accs.append(np.sum(y_pred_label == y_val_label) / 1.0)

print("Accuracy: ", np.mean(accs))
```
